```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 12, 300)
x2 = x[x > 0]

fig = plt.figure(figsize=(3,3), dpi=200)
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel('x')
ax.set_ylabel('y', rotation=0)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

ax.set_xlim([-1.2, 11.2])
ax.set_ylim([-0.4, 10.3])
ax.set_aspect('equal')
ax.set_xticks(range(0, 12, 2))
ax.set_yticks(range(0, 12, 2))
ax.set_xticks(range(-1, 12, 1), True)
ax.set_yticks(range(0, 10, 1), True)

ax.plot(x, np.exp(x))
ax.plot(x2, np.power(x2, np.e))
ax.plot(x2, x2 * np.log(x2))
ax.plot(x, x)
ax.plot(x2, np.log(x2))

ax.text(1.2, 8.8, "$e^n$")
ax.text(2.4, 8.8, "$n^e$")
ax.text(5.2, 8.0, "$n\ ln(n)$")
ax.text(6.8, 6.2, "$n$")
ax.text(8.4, 2.6, "$ln(n)$")

fig.savefig("prog045.png", format="png")
```
